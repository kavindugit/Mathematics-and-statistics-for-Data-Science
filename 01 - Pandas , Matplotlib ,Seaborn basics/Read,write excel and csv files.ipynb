# Read Excel Files

```python
path_excel = r"C:\Users\HP\OneDrive\Desktop\data science\Math -Stats course\3_read_write_to_excel\movies_db.xlsx"
df_movies = pd.read_excel(path_excel , "movies")                   # movies is sheet name in excel file
df_movies.head(4)                                             # in cmd pip install openpyxl
```

```python
df_financials = pd.read_excel(path_excel , "financials")                   
df_financials.head(4)               
```

```python
def standardize_currency(curr):
    if curr == "$$" or curr == "Dollars" :
        return "USD"
    return curr

df_financials = pd.read_excel(path_excel , "financials" , converters = {
    'currency' : standardize_currency
})                   
df_financials.head(4)    




  #    Pandas reads the Excel file, focusing on the specified worksheet "financials".
  #    For the 'currency' column, it sees that you want to use the standardize_currency function.
  #   For each cell in the 'currency' column, it performs something similar to this pseudo-code:

                      #    for cell_value in currency_column:
                      #    new_value = standardize_currency(cell_value)
                      #    insert new_value into DataFrame
```

```python
df_merged = pd.merge(df_movies , df_financials , on ="movie_id")             # merge to data frames on movie id
df_merged.head(4)
```

```python
df_merged.to_excel("movies_merged.xlsx", sheet_name = "merged" , index = False)
```

```python
df_weather = pd.DataFrame({
    'day' :['1/2/2021','12/5/2021','8/4/2021'],
    'temperature' : [32,35,28],
    'event' : ['rain','sunny','snow']
})
df_weather
```

```python
df_stocks = pd.DataFrame({
    'tickers':['Google','WMT','MSFT'],
    'price' :[845,65,64],
    'pe' : [30.37,14.26,30.97],
    'eps' :[27.82 , 4.61 , 2.12]
})
df_stocks
```

```python
with pd.ExcelWriter("stocks_weather.xlsx") as writer:
    df_stocks.to_excel(writer , sheet_name ="stocks")           #   stocks data frame and weather data frame
    df_weather.to_excel(writer , sheet_name ="weather")         #                   export to one exel file as two sheets
```
