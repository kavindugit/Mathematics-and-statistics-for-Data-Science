```python
df["Age"] = df["release_year"].apply(lambda x: 2023 - x)      #create column age using release_year column
df.head(4)                                   # not directly assign release_year , apply a function and assign after execute it
```

```python
df["profit"] = df["revenue"] - df["budget"].apply(lambda x : x + 100)        # this apply function add 100 to budget column 
df.head()                                                                              # before substract from revenue
```

```python
df["profit"] = df["revenue"] - df["budget"]          #create column profit using two available columns in data frame
df.head(4)   
```

```python
df["profit"] = df.apply(lambda x : x["revenue"] - x["budget"] ,axis = 1)          # axis = 0   -> Apply functon to each column
df.head(3)                                                                        # axis = 1   -> Apply function to each row


# how work this code - this code segment use lambda function in apply , the apply fuction assign to whole data frame, 
# in lambda function x is equal to df
```
