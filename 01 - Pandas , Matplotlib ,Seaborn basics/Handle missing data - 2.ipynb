```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\HP\Maths and statistics course codebasics\files\pandas seaborn downdload files\5_handling_missing_data_replace\weather_data.csv")
df
```

```python
df.replace(-99999,0)                             # replace -99999 by 0
```

```python
df.replace(-99999 , np.nan)                            # (value in df , replaced value)
```

```python
df.replace([-99999, -88888] , np.nan)             # use list of values
                                                     # both -99999 and -88888 are replaced by NaN
```

```python
df.replace(to_replace = [-99999, -88888] , value = np.nan)            #same thing like above 
```
