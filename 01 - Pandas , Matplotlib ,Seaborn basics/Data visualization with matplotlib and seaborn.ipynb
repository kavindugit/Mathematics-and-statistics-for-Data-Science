```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_sales = pd.read_excel(r"C:\Users\HP\Maths and statistics course codebasics\files\pandas seaborn downdload files\8_matplotlib_seaborn\linechart.xlsx")
df_sales.head()
```

```python
plt.figure(figsize = (12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"] ,color ="blue" , label = "Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"], color ="green" , label = "Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"] , color ="orange" , label = "Washing Machine")

plt.title("Product Sales")
plt.ylabel("Revenue (min $)")
plt.xlabel("Financial Quater")
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.histplot(df_score["Exam_Score"] , kde = True)      #  kde -> add distribution line to  char

```

```python
sns.histplot(data = df_score , x = "Exam_Score", kde = True) 
plt.xlabel("score")
plt.ylabel("count")
plt.title("Exam Scores")
plt.show()
```
