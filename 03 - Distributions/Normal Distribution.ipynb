```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HP\Maths and statistics course codebasics\files\chapter6_assets\3_normal_distribution\heights.csv")
df.head(3)
```

```python
df.shape
```

```python
mean = round(df.height.mean(),2)
std = round(df.height.std(),2)

mean , std
```

```python
sns.histplot(df.height , kde = True)          # kde use to get distribution line

plt.axvline(mean , color ="black" , linestyle = "dashed" , linewidth = 1)

plt.axvline(mean + std , color ="red" , linestyle = "dashed" , linewidth = 1)
plt.axvline(mean - std , color ="red" , linestyle = "dashed" , linewidth = 1)

plt.axvline(mean + std * 2 , color ="green" , linestyle = "dashed" , linewidth = 1)
plt.axvline(mean - std * 2 , color ="green" , linestyle = "dashed" , linewidth = 1)

plt.axvline(mean + std * 3 , color ="blue" , linestyle = "dashed" , linewidth = 1)
plt.axvline(mean - std * 3 , color ="blue" , linestyle = "dashed" , linewidth = 1)
```

```python
left1 = round(mean - std,2)
right1 = round(mean + std,2)                         #           left3 left2 left1 | right1 right2 right3
left1 , right1
```

```python
left2 = round(mean - std * 2 ,2)
right2 = round(mean + std * 2 ,2)
left2 , right2
```

```python
left3 = round(mean - std * 3 ,2)
right3 = round(mean + std * 3 ,2)
left3 , right3
```

```python
df[(df.height > left1)&(df.height < right1)].shape
```
