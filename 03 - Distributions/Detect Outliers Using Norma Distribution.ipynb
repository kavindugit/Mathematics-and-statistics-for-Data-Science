```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\HP\Maths and statistics course codebasics\files\chapter6_assets\4_detect_outliers_using_normal_distribution\heights_with_outliers.csv")
df
```

```python
mean = df.height.mean()
std = df.height.std()

mean , std
```

```python
sns.histplot(df.height , kde = True)

plt.axvline(mean , color = "k" , linestyle = 'dashed' , linewidth = 1)

plt.axvline(mean + std , color = "r" , linestyle ='dashed' , linewidth = 1)
plt.axvline(mean - std , color = "r" , linestyle ='dashed' , linewidth = 1)

plt.axvline(mean + std * 2 , color = "r" , linestyle ='dashed' , linewidth = 1)
plt.axvline(mean - std * 2 , color = "r" , linestyle ='dashed' , linewidth = 1)

plt.axvline(mean + std * 3 , color = "g" , linestyle ='dashed' , linewidth = 1)
plt.axvline(mean - std * 3 , color = "g" , linestyle ='dashed' , linewidth = 1)
```

```python
lower = mean - 3 * std
upper = mean + 3 * std

lower , upper
```

```python
df[(df.height < lower )| (df.height > upper)]
```

```python
df_no_outliers = df[(df.height > lower) & (df.height<upper)]
df_no_outliers.shape
```

```python
sns.histplot(df_no_outliers , kde = True)
```
