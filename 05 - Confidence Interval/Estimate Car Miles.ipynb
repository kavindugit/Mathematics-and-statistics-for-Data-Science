```python
mean_population = df.miles.mean()
std_dev_population = df.miles.std()

mean_population , std_dev_population

# we goes to pridict that popultion mean using sample 50 data 
# actually in real we didnt have population mean we only have 
#    standard deviation and sample data 
```

```python
SAMPLE_SIZE = 50

sample = df.sample(SAMPLE_SIZE , random_state = 100)
sample.head()
#random_state give us same ramdom value set every times we run this code segment
```

```python
mean_sample = sample.miles.mean()
mean_sample
```

```python
import scipy.stats as st
```

confidece interval is 95%       (given)

area left = (o.95 + 1)/2 = 0.975

now we goes to calculate Z score of that area ......

```python
z_score_95_confidence = st.norm.ppf(0.975)
```

```python
z_score_95_confidence
```

```python
margin_of_error_95 = z_score_95_confidence * (std_dev_population/np.sqrt(SAMPLE_SIZE))
margin_of_error_95
```

```python
lower_95 = mean_sample - margin_of_error_95
upper_95 = mean_sample + margin_of_error_95

lower_95 , upper_95
```

confidece interval is 99% (given)

area left = (o.99 + 1)/2 = 0.995

now we goes to calculate Z score of that area ......


```python
z_score_99_confidence = st.norm.ppf(0.995)
z_score_99_confidence
```

```python
margin_of_error_99 = z_score_99_confidence * (std_dev_population / np.sqrt(SAMPLE_SIZE))

lower_99 = mean_sample - margin_of_error_99
upper_99 = mean_sample + margin_of_error_99

lower_99 , upper_99
```
